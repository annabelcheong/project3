```python
import pandas as pd
from tensorflow.keras.utils import to_categorical # Used for One-hot encoding

from tensorflow.keras.models import Sequential # Produce the deep learning model
from tensorflow.keras.layers import Dense # Layers of model

import joblib # To save model
```

```python
# Retrieve variables from other jupyter notebook
%store -r X_train_scaled
%store -r X_test_scaled
%store -r y_train
%store -r y_test
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)

y_train_categorical.shape
```

```python
# Model Sequential()
# Create a normal neural network with  7 inputs, 4 hidden nodes, and 9 outputs
model = Sequential()
model.add(Dense(units=14, activation='relu', input_dim=7))
model.add(Dense(units=14, activation='relu'))
model.add(Dense(units=14, activation='relu'))
model.add(Dense(units=14, activation='relu'))
model.add(Dense(units=14, activation='relu'))
model.add(Dense(units=9, activation='softmax'))
```

```python
# Summary of model
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=500,
    shuffle=True,
    verbose=2
)
```
